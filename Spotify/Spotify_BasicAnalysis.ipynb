```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Get the files
data = pd.read_excel(r'top50.xlsx')
#test = pd.read_csv('test_V2.csv')
```

```python
data.drop(['Unnamed: 0'], inplace=True, axis=1)
```

Best Song based on Popularity

```python
data.sort_values('Popularity', ascending=False).head(1)['Track.Name']
```

Best Artist based on Popularity

```python
data.sort_values('Popularity', ascending=False).head(1)['Artist.Name']
```

Best Genre based on Popularity

```python
data.sort_values('Popularity', ascending=False).head(1)['Genre']
```

Most Common Genre

```python
data['Genre'].mode()
```

Song - Maximum Length

```python
data[data['Length.'] == max(data['Length.'])][['Track.Name','Artist.Name']] # in seconds
```

Song - Minimum Length

```python
data[data['Length.'] == min(data['Length.'])][['Track.Name','Artist.Name']] # in seconds
```

Find the most popular music according to Genre

```python
data.sort_values(['Popularity','Genre'], ascending=False).drop_duplicates('Genre').set_index('Genre')['Track.Name']
```

Word Cloud of Genre

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                min_font_size = 10).generate(str(data['Genre'].to_list())) 
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wordcloud)
```
